```python
from spectralradex import radex
import numpy as np
from astropy import units as u 
from astropy import constants as c
from astropy.table import Table
import matplotlib.pyplot as plt
import plotly
import plotly.express as px
import matplotlib as mpl
import warnings
import matplotlib.ticker as ticker
from matplotlib.ticker import AutoMinorLocator
warnings.filterwarnings('ignore')
plt.style.use('paper.mplstyle')
```

```python
table_orbit = Table.read('./../data/orbit/orbit_KDL15.dat', format='ascii')
```

```python
table_hncofits = Table.read('./../data/mopra_fits/final_fits_updated.dat', format='ascii')
colnames_new = ['n_gauss', 'x', 'y', 'Ipeak', 'Ipeak_err', 'Vcent', 'Vcent_err', 'FWHM', 'FWHM_err', 'rms', 'residual', 'chisquared', 'DoF', 'reduced_chisquared', 'AIC']

for colname, colname_new in zip(table_hncofits.colnames, colnames_new): 
    table_hncofits[colname].name = colname_new
table_hncofits
```

```python
xmask = (table_hncofits['x']>0.7634) & (table_hncofits['x']<0.859)
ymask = (table_hncofits['y']>-0.2361) & (table_hncofits['y']<-0.1384)
vmask = (table_hncofits['Vcent']>0) & (table_hncofits['Vcent']<70)
table_hncofits_masked = table_hncofits[(xmask&ymask&vmask)]
table_hncofits_masked
```

```python
xcentre = 0.8041918903
ycentre = -0.1838483592
# r_inner = 79/3600
r_inner = 0
r_outer = 154/3600

xmask = (table_hncofits['x']>(xcentre-r_outer)) & (table_hncofits['x']<(xcentre+r_outer))
ymask = (table_hncofits['y']>(ycentre-r_outer)) & (table_hncofits['y']<(ycentre+r_outer))

dist_from_center = np.sqrt((table_hncofits['x'] - xcentre)**2 + (table_hncofits['y'] - ycentre)**2)
circular_mask = (dist_from_center >= r_inner) & (dist_from_center <= r_outer)

vmask_v1 = (table_hncofits['Vcent']>-40) & (table_hncofits['Vcent']<-0)
vmask_v2 = (table_hncofits['Vcent']>0) & (table_hncofits['Vcent']<70)
vmask_v3 = (table_hncofits['Vcent']>70) & (table_hncofits['Vcent']<120)
table_hncofits_masked_v1 = table_hncofits[(circular_mask&vmask_v1)]
table_hncofits_masked_v2 = table_hncofits[(circular_mask&vmask_v2)]
table_hncofits_masked_v3 = table_hncofits[(circular_mask&vmask_v3)]
```

```python
fig, ax = plt.subplots(figsize=(10, 5))

s = (table_hncofits['FWHM']*0.1)
c = (table_hncofits['Ipeak']**0.05)

ax.scatter(table_hncofits['x'], table_hncofits['Vcent'], s=s, c=c, cmap='Greys', alpha=0.75)

ax.scatter(table_hncofits_masked_v1['x'], table_hncofits_masked_v1['Vcent'], s=30, c='Black', alpha=1)
ax.scatter(table_hncofits_masked_v2['x'], table_hncofits_masked_v2['Vcent'], s=30, c='Black', alpha=1)
ax.scatter(table_hncofits_masked_v3['x'], table_hncofits_masked_v3['Vcent'], s=30, c='Black', alpha=1)
ax.scatter(table_hncofits_masked_v1['x'], table_hncofits_masked_v1['Vcent'], s=10, c='C2', alpha=1)
ax.scatter(table_hncofits_masked_v2['x'], table_hncofits_masked_v2['Vcent'], s=10, c='C1', alpha=1)
ax.scatter(table_hncofits_masked_v3['x'], table_hncofits_masked_v3['Vcent'], s=10, c='C0', alpha=1)

p1 = ax.scatter([0], [0], s=30, fc='C2', ec='k', alpha=1, label='<0km/s')
p2 = ax.scatter([0], [0], s=30, fc='C1', ec='k', alpha=1, label='0-70km/s')
p3 = ax.scatter([0], [0], s=30, fc='C0', ec='k', alpha=1, label='>70km/s')

l1 = ax.plot(table_orbit['col10'], table_orbit['col12'], c='k', ls='--', label='KDL15 orbit model', lw=2)

ax.set_ylabel("Velocity [km/s]")
ax.set_xlabel("Galactic Longitude")
ax.set_aspect(0.004615, adjustable='box')
ax.xaxis.set_major_formatter(ticker.StrMethodFormatter("{x:.2f}$^\degree$"))
minor_locator = AutoMinorLocator(5)
ax.xaxis.set_minor_locator(minor_locator)
minor_locator = AutoMinorLocator(5)
ax.yaxis.set_minor_locator(minor_locator)

ax.grid(True, ls=':', alpha=0.25)
for axis in fig.get_axes():
    axis.tick_params(axis='y', labelrotation=90, pad=1)
for tick in ax.yaxis.get_majorticklabels():
    tick.set_verticalalignment("center")

ax.legend(loc='lower right')

first_legend = ax.legend(handles=[p1, p2, p3], loc='lower left')
ax.add_artist(first_legend)
ax.legend(handles=l1, loc='lower right')
ax.set_aspect(3.866e-3, adjustable='box')

ax.set_xlim(1.1,0.075)
ax.set_ylim(-30,125)

fig.savefig('../figures/hnco_lv.pdf', bbox_inches='tight', dpi=200, transparent=False)
```
