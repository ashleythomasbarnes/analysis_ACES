```python
import scipy
import pylab
import numpy as np
import matplotlib
import astropy.io.fits as fits

from matplotlib.colors import ListedColormap
```

```python
#Make layered image

starfile = 'Region7_190_cutout.fits'
stars = fits.open(starfile,ignore_missing_end=True)
gasfile = 'HNC.masked.integrated.reproject.fits'
gas = fits.open(gasfile)
hiifile = 'Region7_palpha_cutout.fits'
hii = fits.open(hiifile,ignore_missing_end=True)

#get color map; in general want something that has a pretty simply monotonic increase starting from black at low values.
cmap = pylab.cm.viridis #pylab.cm.copper # copper can also be a good choice
my_cmap = cmap(np.arange(cmap.N))
# Set alpha
my_cmap[:,-1] = np.linspace(0, 1, cmap.N)
# Create new colormap
my_overlay = ListedColormap(my_cmap)

pink = pylab.cm.pink
my_pink = pink(np.arange(pink.N)) 
my_pink[:,-1] = np.linspace(0, 1, pink.N)
my_pink = ListedColormap(my_pink)

pylab.figure(1,figsize=(15,15))
pylab.imshow(np.flipud(stars[0].data),vmax=5,vmin=0,cmap='bone')
pylab.imshow(np.flipud(hii[0].data/np.nanmax(hii[0].data)),cmap=my_pink) #dumb way to boost intensity of transparent (faint) image: just overplot it a lot. 
pylab.imshow(np.flipud(hii[0].data/np.nanmax(hii[0].data)),cmap=my_pink)
pylab.imshow(np.flipud(hii[0].data/np.nanmax(hii[0].data)),cmap=my_pink)
pylab.imshow(np.flipud(hii[0].data/np.nanmax(hii[0].data)),cmap=my_pink)
pylab.imshow(np.flipud(hii[0].data/np.nanmax(hii[0].data)),cmap=my_pink)
pylab.imshow(np.flipud((gas[0].data)/np.nanmax(gas[0].data)),cmap=my_overlay)
pylab.imshow(np.flipud(gas[0].data/np.nanmax(gas[0].data)),cmap=my_overlay)

pylab.savefig('CND_overlay_v4.pdf')
```
